```python
#범례 표시

data_dict = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
plt.plot('data_x','data_y',data=data_dict,label='Price($)')
plt.xlabel('X축',fontdict={'color':'b','size':'x-large'})
plt.ylabel('Y축',fontdict={'color':'r', 'size':12}) #원래 한글은 깨짐
plt.legend() #범례 표시
plt.show()
```

```python
#범례 위치

data_dict = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
plt.plot('data_x','data_y',data=data_dict,label='Price($)')
plt.xlabel('X축',fontdict={'color':'b','size':'x-large'})
plt.ylabel('Y축',fontdict={'color':'r', 'size':12}) #원래 한글은 깨짐
plt.legend(loc=10) #범례 표시  1 우상 2 좌상 3 좌하.....10 중앙
plt.show()
```

```python
data_1 = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
data_2 = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y',data=data_1,label='Price($)')
plt.plot('data_x','data_y',data=data_2,label='Demand (#)')

plt.xlabel('X축',fontdict={'color':'b','size':'x-large'})
plt.ylabel('Y축',fontdict={'color':'r', 'size':12})
#plt.legend() #범례 표시는 1열이 기본
#plt.legend(ncol=2) #ncol 옵션을 사용하여 열이 2개로 표현
#plt.legend(fontsize=12, shadow=True) #그림자
#plt.legend(frameon=False) #테두리 없이
plt.show()
```

```python
#축 설정
plt.plot([1,2,3,4],[2,3,5,10])
plt.axis('square') #가로축과 세로축을 같게 ->왜곡을 막을 수 있다
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10])
plt.axis('scaled') #square는 여백이 많이 생기지만 맞춰서 해줌
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
#특별한 설정이 없으면 그냥 실선으로 그래프를 그리게 된다

plt.plot([1,2,3,4],[2,3,5,10],'bo-') #옵션 bo- 주면 블루 지점표시 + 실선

plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.plot([1,2,3,4],[2,3,5,10],'bo--') #옵션 bo- 주면 블루 지점표시 + 점선

plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
#Q)세개의 선을
#빨강 실선 [1,2,3,4],[2,2.5,3.3,4.3] 리플
#그린 점선 [1,2,3,4],[2,3,3,6] 이더리움
#블루 삼각형 [1,2,3,4],[2,3,4,10] 비트코인

data_1 = {'data_x':[1,2,3,4],'data_y':[2,2.5,3.3,4.3]}
data_2 = {'data_x':[1,2,3,4],'data_y':[2,3,3,6]}
data_3 = {'data_x':[1,2,3,4],'data_y':[2,3,4,10]}

plt.plot('data_x','data_y','r-',data=data_1,label='리플')
plt.plot('data_x','data_y','g:',data=data_2,label='이더리움')
plt.plot('data_x','data_y','b^',data=data_3,label='비트코인')
plt.legend(ncol=3)
plt.show()
```

```python
import numpy as np #계산 관련 라이브러리
a=np.arange(0,2,0.2) #0부터 2미만까지 0.2 리스트 생성

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo',markersize=20) #마커의 크기 설정
plt.plot(a,a**3,'g-.',linewidth=10) #선의 두께 설정

# plt.grid(True) #그리드 설정: x,y축 모두 설정
plt.grid(True,axis='y') #그리드 설정,y축만 설정
plt.show()
```

```python
#타이틀 설정

import numpy as np #계산 관련 라이브러리
a=np.arange(0,2,0.2) #0부터 2미만까지 0.2 리스트 생성

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo',markersize=20) #마커의 크기 설정
plt.plot(a,a**3,'g-.',linewidth=10) #선의 두께 설정

# plt.grid(True) #그리드 설정: x,y축 모두 설정
plt.grid(True,axis='y') #그리드 설정,y축만 설정

#plt.title('그래프 제목')
plt.title('그래프 제목',loc='right',fontdict={'fontsize':16,'fontweight':'bold'})
plt.show()
```

```python
#막대 그래프 bar

z=np.arange(3)
x=['2020','2021','2022']
y= [100,400,900]

plt.bar(z,y)
plt.xticks(z,x) #x축의 값을 표현
plt.show()
```

```python
z=np.arange(3) #0,1,2
x=['2020','2021','2022']
y= [100,400,900]

plt.bar(z,y,width=0.6,align='edge',color='green') #옵션 추가
plt.xticks(z,x) 
plt.show()
```

```python
z=np.arange(3) #0,1,2
x=['2020','2021','2022']
y= [100,400,900]

plt.barh(z,y,color='green') #옵션 추가 barh에서 h 붙여서 horizontal
plt.yticks(z,x) 
plt.show()
```

```python
#산정도 scatter

np.random.seed(1) # 난수 발생시 일정한 값으로 난수 발생
n=50

x=np.random.rand(n) #난수를 발생시키는데 0~1 시가의 난수 발생
y=np.random.rand(n)

plt.scatter(x,y)
plt.show()
```

```python
np.random.seed(1) # 난수 발생시 일정한 값으로 난수 발생
n=50

x=np.random.rand(n) #난수를 발생시키는데 0~1 시가의 난수 발생
y=np.random.rand(n)
size=(30 * np.random.rand(n)) **2
color=np.random.rand(n)

plt.scatter(x,y,s=size,c=color, alpha=0.4) #alpha는 투명도
plt.show()
```

```python
np.random.seed(1) # 난수 발생시 일정한 값으로 난수 발생
n=50

x=np.random.rand(n) #난수를 발생시키는데 0~1 시가의 난수 발생
y=np.random.rand(n)
size=(30 * np.random.rand(n)) **2
color=np.random.rand(n)

plt.scatter(x,y,s=size,c=color, alpha=0.4, cmap='Set1') #alpha는 투명도
plt.colorbar()
plt.show()
```
